```python
# Preprocessing on Dataset1
import pandas as pd
data={
    "RollNo":[101,102,103,104,105],
    "Name":["A","B","C","D","E"],
    "Marks":[90,80,70,70,"."],
    "Age":[20,19,18,19,20],
    "Department":["Cse","Mech","Civil","Ece","Mech"],
    "Admission_Date":["01-06-2023","02-03-2023","03-08-2023",".","15-06-2023"]
}
df1=pd.DataFrame(data)

#Save as CSV
df1.to_csv("student_data.csv",index=False)
```

```python
import pandas as pd
import numpy as np

#Step 1: Load the dataset
df=pd.read_csv("student_data.csv")
print("Original Data:")
df
```

```python
# Step 2: Replace "." with NaN to mark missing values
df.replace(".",np.nan,inplace=True)
df
```

```python
#Convert columns to numeric wherever needed
df["Age"]=pd.to_numeric(df["Age"],errors="coerce")
df["Marks"]=pd.to_numeric(df["Marks"],errors="coerce")
df

'''When converting data (like strings to datetime),
if some values are invalidor can not be converted,
instead of throwing an error,it will:
replace those invalid values with NaT(Not a Time)'''
df
```

```python
#Step 3: Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Step 4: Fill missing Age with mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df
```

```python
#Step 5: Fix inconsistent text formatting
df["Department"]=df["Department"].str.upper()
df
```

```python
print(df.isnull().sum())
```

```python
#Step 6: Convert Admission Date to standard Datetime format
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"],errors="coerce")
df
```

```python
# Step 7: Show Cleaned Data
print("\nCleaned Data")
df
```

```python
#Step 8: Save Cleaned Dataset
df.to_csv("Cleaned_Student_data.csv",index=False)
print("\nCleaned CSV file saved as cleaned_student_data.csv")
```

```python
# Working with missing values on Dataset 2
df=pd.read_csv('/content/Cars93.csv')
df
```

```python
df.isna().sum()
```

```python
# Imputing the missing value
# 1. Lugguage rooom
df['Luggage.room'].fillna(df['Luggage.room'].mean(),inplace=True)
df
```

```python
# 2. Rear Seat Room
df["Rear.seat.room"]=df["Rear.seat.room"].fillna(value=df["Rear.seat.room"].mean())
df
```

```python
# 3. Airbag
df["AirBags"]=df["AirBags"].fillna(value=df["AirBags"].mode().values[0])
df
```

```python
# Save Cleaned Dataset
df.to_csv("Cleaned_Cars93.csv",index=False)
```
