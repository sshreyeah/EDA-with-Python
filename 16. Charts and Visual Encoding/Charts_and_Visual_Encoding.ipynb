```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd
data={'Days':['Mon','Tue','Wed','Thur','Fri'],
      'study_hours':[2,6,4,9,6],
      'Marks':[50,60,70,60,80],
      'Attendace':[60,70,87,89,90],
      'Sleep_hours':[6,8,9,10,6],
      'Assignment_Completed':[1,3,4,5,6]}
df=pd.DataFrame(data)
df
```

```python
# 1. Line Chart
plt.plot(df['Days'],df['study_hours'],marker='o')
plt.title('Study hours over days')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Advanced Line chart
plt.figure(figsize=(10,6))
plt.plot(df['Days'],df['Marks'],marker='o',label='Marks',color='lightsalmon')
plt.plot(df['Days'],df['study_hours'],marker='s',label='study_hours',color='cyan')
plt.title('Marks and study_hours over days')
plt.xlabel('Days')
plt.ylabel('Marks and study_hours')
plt.legend()
plt.show()
```

```python
# 2. Bar chart
plt.bar(df['Days'],df['Marks'],color='steelblue')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
# Advanced Bar Graph
bars=plt.bar(df['Days'],df['Marks'],color='orchid')
for bar in bars:
  y=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,y,str(y),ha='center',va='bottom')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(True) #adds horizontal grid for readability
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(len(df['Marks']))
width=0.25
plt.bar(x-width/2,df['Marks'],width,label='study_hours')
plt.bar(x+width/2,df['study_hours'],width,label='Marks')
plt.xticks(x,df['Marks'])
plt.xlabel('Marks')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# 3. Histogram
# Shows how values are spread
data={'Marks':[50,60,70,70,80,90,70,40,50,60,75,88,87,67,56]}
df=pd.DataFrame(data)
plt.hist(df['Marks'],bins=5,edgecolor='black',alpha=0.5)
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Scatter Plot (Relationship)
import matplotlib.pyplot as plt
plt.scatter(df['study_hours'],df['Marks'])
plt.title("Study Hours VS Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
import pandas as pd
data={'Days':['Mon','Tue','Wed','Thur','Fri'],
      'study_hours':[2,6,4,9,6],
      'Marks':[50,60,70,60,80],
      'Attendace':[60,70,87,89,90],
      'Sleep_hours':[6,8,9,10,6],
      'Assignment_Completed':[1,3,4,5,6]}
df=pd.DataFrame(data)
# Conditional Scatter Plot
# Example category
df['Result']=['Pass','Fail','Pass','Fail','Pass']
colors=['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['study_hours'],df['Marks'],c=colors)
plt.title("Study Hours VS Marks (Result-wise)")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

# Same Graphs using Seaborn (Better Visualization)

```python
data = {
    "Day": ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun', 'Mon', 'Tue', 'Wed'],
    "Region": ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    "Sales": [200, 220, 250, 270, 300, 350, 400, 280, 260, 290],
    "Profit": [20, 25, 30, 35, 50, 60, 80, 40, 38, 45],
    "Customers": [30, 35, 40, 45, 50, 65, 80, 55, 48, 52],
    "Category": ['Food', 'Clothing', 'Electronics', 'Food', 'Clothing', 'Electronics', 'Food', 'Clothing', 'Electronics', 'Food']
}
df_new = pd.DataFrame(data)
df_new
```

```python
# Line Chart
sns.lineplot(data=df_new,x='Day',y='Sales')
plt.title('Sales Over Days')
plt.show()
```

```python
# Advanced Line Chart
sns.lineplot(data=df_new,x='Day',y='Profit',hue='Region')
plt.title('Profit Over Days (Region-wise)')
plt.show()
```

```python
# Bar Graph
sns.barplot(data=df_new,x='Region',y='Sales')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales by Region')
plt.show()
```

```python
# Advanced Bar Graph
sns.barplot(data=df_new,x='Region',y='Profit',hue='Category')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Region and Category')
plt.show()
```

```python
# Histogram
sns.histplot(data=df_new,x='Profit',bins=5,edgecolor='black',alpha=0.5)
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
# Advanced Histogram
sns.histplot(data=df_new,x='Customers',bins=5)
sns.histplot(data=df_new,x='Sales',bins=5)
plt.title('Customers and Sales Distribution')
plt.xlabel('Customers/Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(data=df_new,x='Customers',y='Sales',hue='Category')
plt.xlabel('Customers')
plt.ylabel('Sales')
plt.title('Sales vs Customers (Category-wise)')
plt.show()
```

```python
# Advanced Scatter Plot
sns.scatterplot(data=df_new,x='Customers',y='Profit',hue='Region')
plt.xlabel('Customers')
plt.ylabel('Profit')
plt.title('Profit vs Customers (Region-wise)')
plt.show()
```
