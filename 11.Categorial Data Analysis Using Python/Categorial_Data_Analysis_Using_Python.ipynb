```python
import pandas as pd
data={
'Order_ID':['01','02','03','04','05','06','07','08','09','010'],
       'Category':['Furniture','Office Supplies','Furniture','Technology','Technology','Furniture','Office Supplies','Furniture','Technology','Technology'],
       'Payment_Mode':['COD','COD','UPI','COD','UPI','COD','COD','UPI','COD','UPI'],
       'Delivery_Type':['Express','Standard','Express','Standard','Express','Standard','Express','Standard','Express','Standard'],
       'Customer_Type':['New','Returning','New','Returning','New','New','Returning','New','Returning','New']
}
df=pd.DataFrame(data)
df
```

```python
#Frequency Count
#Count how many orders in each category
print(df['Category'].value_counts())
print( df['Payment_Mode'].value_counts())
print(df['Delivery_Type'].value_counts())
```

```python
#Unique Values
#Display all unique categories
print(df['Category'].unique())
print(df['Payment_Mode'].unique())
print(df['Delivery_Type'].unique())
print(df['Customer_Type'].unique())
```

```python
#Number of Unique Values
print(df['Category'].nunique())
print(df['Payment_Mode'].nunique())
print(df['Delivery_Type'].nunique())
print(df['Customer_Type'].nunique())
```

```python
#Cross Tabulation
#Category vs Payment Method
print(pd.crosstab(df['Category'],df['Payment_Mode']))
```

```python
#Cross Tabulation
#Delivery Type vs Customer Type
print(pd.crosstab(df['Delivery_Type'],df['Customer_Type']))
```

```python
#Percentage Distribution
print(df['Category'].value_counts(normalize=True)*100)
```

```python
#Filtering Data
#Orders from Electronics Category
print(df[df['Category']=='Furniture'])
```

```python
#Grouping
#Orders by Category and Payment Method
print(df.groupby('Category')['Payment_Mode'].value_counts())
```

```python
#Sorting Data
sorted_df=df.sort_values(by='Category')
print(sorted_df)
```
