#**Part A: Data Normalization**

```python
#Dataset
import pandas as pd
import numpy as np

data= {'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
       'Price':[55000,30000,67000,30000,55600,45000],
       'Units_sold':[25,45,67,89,65,34],
       'Discount':[10,20,15,25,10,5]}
df=pd.DataFrame(data)
print(df)
```

```python
#1. Min-Max Normalization
#Min-Max Normalization for Price

df['Price_MinMax']=(df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())
print("\nMin-Max Normalized price:")
print(df[['Product','Price','Price_MinMax']])
```

```python
df['Discount_MinMax']=(df['Discount']-df['Discount'].min())/(df['Discount'].max()-df['Discount'].min())
print("\nMin-Max Normalized prDiscountce:")
print(df[['Product','Discount','Discount_MinMax']])
```

```python
#Normalized Multiple Columns
cols=['Price','Units_sold','Discount']
df_norm=(df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
print("\nMin-Max Normalized Data:")
df[cols] = df_norm
df_norm
```

```python
#2. Z-Score Normalization (Standardization)
df['Units_Zscore']=df['Units_sold']-df['Units_sold'].mean()/df['Units_sold'].std()
print("\nZ-Score Normalized Data:")
print(df[['Product','Units_sold','Units_Zscore']])
```

```python
#3. Decimal Scaling
df['Price_Decimal']=df['Price']/100000
print("\nDecimal Scaled Data:")
print(df[['Product','Price','Price_Decimal']])
```

#**Part B: Turning Categorical Variables into Quantitative Variables in Python**

```python
data={
    'Order_ID':[101,102,103,104,105,106],
    'Customer_Gender':['Male','Female','Male','Female','Male','Female'],
    'Payment_Mode':['Credit Card','COD','Debit Card','COD','Credit Card','COD'],
    'Product_Category':['Electronics','Clothing','Electronics','Clothing','Electronics','Clothing'],
    'City':['New York','Los Angeles','New York','Los Angeles','New York','Los Angeles'],
    'Oder_Value':['4500','2300','3200','5200','1800','4100']
}
df=pd.DataFrame(data)
df
```

```python
#Method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder ()
df['Customer_Gender_Encoded']=le.fit_transform(df['Customer_Gender'])
print (df)
```

```python
#Method 2: One-Hot Encoding
df_encoded=pd.get_dummies(df,columns=['Payment_Mode'])
print(df_encoded)
```

```python
#Method 3:Dummy Encoding (Drop First)
df_dummy=pd.get_dummies(df,columns=['Product_Category'],drop_first=True)
print(df_dummy)
```
